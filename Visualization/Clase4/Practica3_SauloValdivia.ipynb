```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Grafo de Generos Musicales
Es un grafo que indica los diferentes generos musicales y como evolucionaron en otros generos musicales derivados. Las aristas muestran ejemplos de grupos o bandas relacionadas que pertenecen a estos generos.

```python
import csv
in_file=csv.reader(open('MusicGenres.csv','r'))

I=nx.Graph()

for line in in_file:
    I.add_edge(line[0],line[1])
    
I.nodes(data=True)
I.edges(data=True)
```

```python
plt.rcParams['figure.figsize'] = (30.0, 20.0)

positions = nx.kamada_kawai_layout(I)

nx.draw(I, pos=positions,
            node_color='lightblue',
            edge_color='gray',
            font_size=20,
            width=1, with_labels=True, node_size=500, alpha=0.8
       )
```

```python
def DrawSubGenre(genreFile, genreColor, fontColor, positions):
    in_file=csv.reader(open(genreFile,'r'))

    F=nx.Graph()

    for line in in_file:
        F.add_edge(line[0],line[1], label = line[2])

    nx.draw(F, pos=positions,
                node_color=genreColor,
                edge_color=genreColor,
                font_size=20,
                width=3, with_labels=True, node_size=500, alpha=0.8
           )

    for e in F.edges(data=True):
        nx.draw_networkx_edge_labels(F,
                                     positions,
                                     edge_labels={(e[0], e[1]):F.edges[e[0], e[1]]['label']},
                                     font_size=15,
                                     font_color=fontColor)
```

```python
plt.rcParams['figure.figsize'] = (30.0, 20.0)

positions = nx.kamada_kawai_layout(I)

nx.draw(I, pos=positions,
            node_color='lightblue',
            edge_color='gray',
            font_size=20,
            width=1, with_labels=True, node_size=500, alpha=0.8
       )

DrawSubGenre('Funk.csv', '#CBF6D9', 'green', positions)
DrawSubGenre('RockMusic.csv', '#808080', 'gray', positions)
DrawSubGenre('HardRock.csv', '#EEF6C4', 'black', positions)
DrawSubGenre('AlternativeRock.csv', 'lightblue', 'black', positions)
DrawSubGenre('ProgressiveRock.csv', '#C455DC', 'black', positions)
DrawSubGenre('Electronica.csv', '#568BE6', 'blue', positions)
DrawSubGenre('Jazz.csv', '#E99899', 'red', positions)
DrawSubGenre('Soundtrack.csv', '#FF8409', 'orange', positions)
```
