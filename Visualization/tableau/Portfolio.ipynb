```python
import pandas as pd
```

```python
df = pd.read_csv('P1.csv')
df.head()
```

```python
import yfinance as yf  

data = yf.download('AAPL','2019-10-31','2019-10-31')
print(data['Close'][0])

prices = []

for index, row in df.iterrows():
    print(row['Security'] + ' ' + row['Date'])    
    data = yf.download(row['Security'],row['Date'],row['Date'])    
    if data.empty:
        #print("An exception occurred")
        prices.append(0)
    else:
        #print(data)
        prices.append(round(data['Close'][0], 2))
```

```python
#print(type(prices))
df['Price'] = prices

df['Total'] = df['Price'] * df['Quantity']
```

```python
df= df[df['Price'] != 0]
df
```

```python
df = df.round(2)
df.to_csv('Portfolio2.csv', index = False)
```
