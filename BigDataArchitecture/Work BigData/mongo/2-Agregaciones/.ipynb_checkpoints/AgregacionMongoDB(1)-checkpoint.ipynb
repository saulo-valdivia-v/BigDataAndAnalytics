```python
import os
from pymongo import MongoClient
```

```python
ip = "localhost"
```

```python
def db_connection(ip, db_name, collection_name):
    url = 'mongodb://'+ip+':27017/'
    print(url)
    client = MongoClient(url)
    db = client[db_name]
    collection = db[collection_name]
    print('db {} connected, collection {} current data: {}'.
           format(db_name, collection_name, collection.count()))
    return collection

def db_curso_zips():
    db_name = "Curso"
    collection_name = "zips"
    zips = db_connection(ip ,db_name, collection_name)
    return zips
```

```python
zips = db_curso_zips()
```

### Operador $Match

```python
pipeline = [
    {"$match" : {"state" : "CA"} }
]

cursor = zips.aggregate(pipeline)
list(cursor)[:5]
```

```python
#$sample Randomly selects the specified number of documents from its input.
pipeline = [
    {"$match" : {"state" : "CA"} },{"$sample":{"size":2}},
]

cursor = zips.aggregate(pipeline)
list(cursor)
```

```python
# Return States with Populations above 10 Million
pipeline = [    
    { "$group": { "_id": "$state", "totalPop": { "$sum": "$pop" } } },
    { "$match": { "totalPop": { "$gte": 10*1e6 } } }
] 

cursor = zips.aggregate(pipeline)
list(cursor)
```

```python
pipeline =  [ 
        {  "$group" : {  
            "_id" : "$state" ,  
            "poblacionMedia" : {"$avg" : "$pop"}  
            }
        },{
            "$limit":10
        }
]

cursor = zips.aggregate(pipeline)
list(cursor)
```

The equivalent SQL for this aggregation operation is:

SELECT state, SUM(pop) AS totalPop
FROM zipcodes
GROUP BY state
HAVING totalPop >= (10*1000*1000)

```python
pipeline = [ 
    {  
    "$group" : {  
        "_id" : { "estado" : "$state" , "ciudad" : "$city"} ,  
        "poblacion" : {"$sum" : "$pop"}  
    }
} , {    
    "$group" : {
        "_id" : "$_id.estado" ,  
        "poblacionTotal" : {"$avg" : "$poblacion"}  
    }
} ]

cursor = zips.aggregate(pipeline)
list(cursor)[:5]
```

```python
pipeline = [ {  
    "$group" : {  
        "_id" : "$state" ,  
        "poblacionMinima" : {"$min" : "$pop"}  
    }
} ]

cursor = zips.aggregate(pipeline)
list(cursor)[:10]
```

### Operador $project

```python
pipeline = [
    {
        "$project" : {
            "_id" : 0,
            "estado" : "$state" ,  
            "poblacion" : "$pop"  
        } 
    } ,  {
    "$group" : {
        "_id" : "$estado",
        "poblacionTotal" : { "$sum": "$poblacion"}
    }    
    } ]

cursor = zips.aggregate(pipeline)
list(cursor)[:10]
```

```python
pipeline = [ {  
    "$group" : {  
        "_id" : "$state" ,  
        "poblacionMinima" : {"$min" : "$pop"}  
    }
} , { "$project" : { 
            "_id" : 1,
            "poblacionMinima": 1,
            "abandonado" : { "$lt" : ["$poblacionMinima", 1] } } 
        }

]
cursor = zips.aggregate(pipeline)
list(cursor)[:10]
```

#### Y usando el operador out lo guardo en otra colección

```python
pipeline = [
    {  
        "$group" : {  
            "_id" : "$state" ,  
            "poblacionMinima" : {"$min" : "$pop"}  
        }
    } , { "$project" : { 
                "_id" : 1,
                "poblacionMinima": 1,
                "abandonado" : { "$lt" : ["$poblacionMinima", 1] } } 
    },
    {"$out":"ciudades_abandonadas"}
]
cursor = zips.aggregate(pipeline)
```

### Trabajando con cadenas de texto y arrays

```python
pipeline = [ {  
    "$group" : {  
        "_id" : "$state" ,  
        "ciudades" : {"$push" : { "$substr" : [ "$city" , 0 , 5 ] } }  
    }
} , { "$limit" : 1 }  ]

cursor = zips.aggregate(pipeline)

#Del registro que devuelve mostrar las 5 primeras ciudades
doc = dict(cursor.next())
print (doc["_id"], doc["ciudades"][:5])
```

```python
pipeline = [ {  
    "$group" : {  
        "_id" : "$state" ,  
        "ciudades" : {"$push" : { "$toLower" : "$city" } }  
    }
} , { "$limit" : 1 }  ]
cursor = zips.aggregate(pipeline)
print (f'_id:{cursor["_id]"')
list(cursor)
```

### Operadores sort, skip y limit

```python
pipeline = [
    {"$match" : {"state" : "TX"} }, 
    {
        "$group" : {
            "_id" : "$city",
            "poblacion" : { "$sum": "$pop"}
        }    
    },  
    { "$sort" : {"poblacion" : -1} 
    },
    {"$skip":2},{"$limit":4}
]
cursor = zips.aggregate(pipeline)
list(cursor)
```

```python
#Es importante el orden de los operadores
zips.aggregate ( [
    {"$match" : {"state" : "TX"} },
    {
        "$group" : {
            "_id" : "$city",
            "poblacion" : { "$sum": "$pop"}
        }    
    },  
    { "$sort" : {"poblacion" : -1}
    },
    {"$limit":4},{"$skip":2}
        
]    )

cursor = zips.aggregate(pipeline)
list(cursor)
```
