```python
# Propiedades
from numpy.random import randn

datos = randn(30).cumsum()
plt.plot(datos, 'ko--')
```

```python
plt.plot(datos, color='k', linestyle='dashed', marker='o')
```

```python
# Interpolando lineas rectas

plt.plot(datos, 'k--', label='Default')
plt.plot(datos, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
# comparar datos
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(np.random.randn(50).cumsum(), 'k--')
ax2.plot(data)
ax3.plot(data2[0], data2[1], 'ro')
ax4.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
datos2 = np.random.randn(1000).cumsum()
ax.plot(datos2)
```

```python
# Ticks agregados
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(datos2)

ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['uno', 'dos', 'tres', 'cuatro', 'cinco'], rotation =30, fontsize='small')
ax.set_title('titulo')
ax.set_xlabel('escala')
ax.set_ylabel('Amplitud')

ax.plot()
```

```python
# Ticks agregados
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'k', label='uno')
ax.plot(randn(1000).cumsum(), 'b--', label='dos')
ax.plot(randn(1000).cumsum(), 'k.', label='tres', color='red')

ax.legend(loc='best')
```

```python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Renombrar la columna a Region(SouthEast, SouthWest...) en base a State

plt.plot(df['state'], df['beef'], label='Beef')
plt.plot(df['state'], df['pork'], label='Pork')

plt.legend(loc='upper left')
```

```python
# Renombrar la columna Category a Region(SouthEast, SouthWest...) en base a State

plt.plot(df['state'], df['beef'], label='Beef')
plt.plot(df['state'], df['pork'], label='Pork')

plt.legend(loc='upper right')

etiquetas = np.array(df['state'])

plt.xticks(etiquetas, rotation='vertical')

plt.title('Vacuno y Porcino')
plt.ylabel('Produccion')
plt.xlabel('Estado')

plt.show()
```

```python
# Ampliar los graficos
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
plt.plot(df['state'], df['beef'], label='Beef')
plt.plot(df['state'], df['pork'], label='Pork')

plt.legend(loc='upper right')

etiquetas = np.array(df['state'])

plt.xticks(etiquetas, rotation='vertical')

plt.title('Vacuno y Porcino', fontsize = 14)
plt.ylabel('Produccion', fontsize=14, fontfamily='monospace', color='green')
plt.xlabel('Estado')

plt.show()
```
