```python
import pandas as pd
import numpy as np
```

```python
url = 'Ventas.xlsx'
df = pd.read_excel(url)
df.head()
```

```python
# Indices de agrupacion en orden de peticion
# si no se pasa un aggFunction se realiza una media(mean)
print(pd.pivot_table(df,index=['Manager', 'Nombre', 'Rep']))

print(pd.pivot_table(df,index=['Rep', 'Manager', 'Nombre']))
```

```python
pd.pivot_table(df, index=['Manager', 'Nombre', 'Rep'], aggfunc={'Cantidad': np.sum, 'Precio': np.sum})
```

```python
# Especificando los valores y columnas que queremos que sean agregadas
# Cantidad indica la cantidad de filas a agrupar
# Precio realiza la suma del grupo
# Los aggFunc se evaluan a todas las columnas numericas
pd.pivot_table(df, index=['Manager', 'Nombre', 'Rep'], values=['Precio', 'Cantidad'], aggfunc={'Precio': np.sum, 'Cantidad': np.size})
```

```python
# Especificando los valores y columnas que queremos que sean agregadas
# Cantidad indica la cantidad de filas a agrupar
# Precio realiza la suma del grupo
# Los aggFunc se evaluan a todas las columnas numericas
# Multiples aplicaciones a aggfunc
pd.pivot_table(df, index=['Manager', 'Nombre', 'Rep'], values=['Precio', 'Cantidad'], aggfunc=[np.sum, np.size])
```

```python
pd.pivot_table(df, index=['Manager', 'Nombre', 'Rep'], columns=['Producto'], values=['Precio', 'Cantidad'], aggfunc=np.sum)
```

```python
# Margins agrega nueva columna con el total de los calculos de valores
pd.pivot_table(df, index=['Manager', 'Nombre', 'Rep'], columns=['Producto'], values=['Precio'], aggfunc=np.sum, margins='True')
```

```python
# Len permite realizar un recuento
pd.pivot_table(df, index=['Manager', 'Rep'], columns=['Producto'], values=['Precio'], aggfunc=[np.mean, len])
```
